```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(1,11) # 11 is exclusive
x
```

```python
y = x*2
y
```

```python
plt.plot(x,y)
```

 # title xlabel ylabel

```python
plt.plot(x,y)
plt.title("First Line Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

# Changing Line Aesthetics

```python
plt.plot(x,y,color="g",linestyle=":",linewidth=2)
# color="g" = green color
# linestyle=":" = dotted line
plt.show()
```

# Adding 2 lines in a the same plot

```python
y2 = x*3
y2
```

```python
plt.plot(x,y,color="g",linestyle=":",linewidth=2)
plt.plot(x,y2,color="r",linestyle="-.",linewidth=3)
plt.title("2 Line plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True)
plt.show()
```

# adding 2 sub-plot

```python
# plot along columns
plt.subplot(1,2,1)
plt.plot(x,y,color='orange',linewidth=2)

plt.subplot(1,2,2)
plt.plot(x,y2,color='yellow')

plt.show()
```

```python
# plot along rows
plt.subplot(2,1,1)
plt.plot(x,y,color='orange',linewidth=2)

plt.subplot(2,1,2)
plt.plot(x,y2,color='yellow')

plt.show()
```
