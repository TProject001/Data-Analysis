```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Create data
fruit = ["Apple","Mango","Orange","Banana"]
quantity = [30,50,20,60]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
# Add percentage

plt.pie(quantity,labels=fruit,autopct="%0.1f%%")
plt.show()
```

```python
# Change color

plt.pie(quantity,labels=fruit,autopct="%0.1f%%",
       colors=["pink","green","Orange","yellow"])
plt.show()
```

# DoughNut-Chart

```python
plt.pie(quantity,labels=fruit,radius =2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,radius =2)
plt.pie([7],colors=['yellow'],radius=1) # 7 is any random number
plt.show()
```
