```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.random.randint(1,10,7)
x
```

```python
a = np.random.randint(1,10,7)
a
```

```python
plt.scatter(x,a)
plt.show()
```

# Changing Mark Aesthetics

```python
plt.scatter(x,a,marker="*",c='g',s=100)
plt.show()
```

```python
plt.scatter(x,a,c='orange',marker=".",s=500)
plt.show()
```

# Adding 2 marker in the same plot

```python
y2 = np.random.randint(1,10,7)
y2
```

```python
plt.scatter(x,a,c='Orange',s=200)
plt.scatter(x,y2,c="r",marker="*",s=200)

plt.show()
```

# subplot

```python
plt.subplot(1,2,1)
plt.scatter(x,a,c='g',s=100)

plt.subplot(1,2,2)
plt.scatter(x,y2,marker="*",s=100)

plt.show()
```

```python
plt.subplot(2,1,1)
plt.scatter(x,a,c='g',s=100)

plt.subplot(2,1,2)
plt.scatter(x,y2,marker="*",s=100)

plt.show()
```
