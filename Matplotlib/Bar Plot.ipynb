```python
from matplotlib import pyplot as plt
```

```python
Students = {"Bob":87,"Matt":67,"Sam":90}
names = list(Students.keys())
marks = list(Students.values())
```

```python
plt.bar(names,marks)
plt.show()
```

```python
plt.bar(names,marks)
plt.title("Student Marks")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
# horizontal Marks
plt.barh(names,marks)
plt.show()
```
