Import Require packages

```python
#Import Packages
import requests
from bs4 import BeautifulSoup as Soup
import pandas as pd
```

Get Webpage Url

```python
User_Input = input("Enter Product name : ")
User_Input = User_Input.replace(" ","+")
Main_Url = f"https://flipkart.com/search?q={User_Input}"
```

```python
print(Main_Url)
```

```python
req = requests.get(Main_Url)
```

Declare a variable to store product info

```python
# Create a variables to store info
product_names = []
product_prices = []
product_ratings = []
Product_reviews = []
product_page_links=[]
```

```python
Main_Content = Soup(req.content , "html.parser")

# Pages variable store all url link of products page link
All_pages_data = Main_Content.find_all("a",{'class':'ge-49M'})

product_count =0

#pages contain content for page on which product are present
# iterate one page
for Page in All_pages_data:
    if Page.has_attr("href"):  
        Page_URL = Main_Url+""+Page["href"]
        Page_req = requests.get(Page_URL)
        Page_content = Soup(Page_req.content , "html.parser")
        
        # data contain content for one product
        data = Page_content.find_all('div',{'class':'_2kHMtA'})
        
        for item in data:
            #product name
            name = item.find('div',attrs={'class':'_4rR01T'})
            print(name.text)
            product_names.append(name.text)
            
            #price 
            price = item.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
            print(price.text)
            product_prices.append(price.text)

            # rating
            rating = item.find('div',attrs={'class':'_3LWZlK'})
            if rating is None:
                print("Null")
                product_ratings.append("Null")
                #pass
            else:
                print(rating.text)
                product_ratings.append(rating.text)

            # total review
            total_review = item.find('span',attrs={'class':'_2_R_DZ'})
            if total_review is None:
                print("Null")
                Product_reviews.append("Null")
                #pass
            else:
                print(total_review.text)
                Product_reviews.append(total_review.text)
                #pass

            # Product page link
            link = item.find('a')['href']
            print(Main_Url+link)
            product_page_links.append(Main_Url+link)
            
            # Increment product count
            product_count+=1
```

```python
print(product_count)
```

```python
print(product_ratings)
```

Import Data Into CSV files

```python
Whole_data = {"Product_name":product_names ,
              "Product_price":product_prices ,
              "Product_rating":product_ratings ,
              "Product_review":Product_reviews ,
              "Product_page":product_page_links}
```

```python
dataframe = pd.DataFrame(Whole_data)
```

```python
print(dataframe)
```

```python
dataframe.to_csv("DellLaptop2.csv")
```
