# concat

```python
dict = {
    "Cricket" :[1,6,np.nan,4,3],
    "Football":[4,2,np.nan,5,np.nan],
    "Baseball":[2,4,5,7,1]
}
df1= pd.DataFrame(dict)
df1
```

```python
dict2 = {
    "BasketBall" :[8,9,4,2,10],
    "Golf":[1,2,np.nan,7,np.nan],
    "Baseball":[2,4,5,7,1]
}
df2 = pd.DataFrame(dict2)
df2
```

```python
df = pd.concat([df1,df2]) # add horizontally
df
```

```python
df = pd.concat([df1,df2],axis=1) # add vertically
df
```

# merge 2 table

```python
dict2 = {
    "BasketBall" :[8,9,4,2,10],
    "Golf":[1,2,np.nan,7,np.nan],
    "Baseball":[2,4,5,7,10]
}
df2 = pd.DataFrame(dict2)
df2
```

```python
pd.merge(df1,df2,how="outer",on="Baseball")
```

```python
pd.merge(df1,df2,how="inner",on="Baseball")
```

The one difference between merge and join is that,
merge uses common column to combins 2 dataframe,
wheras join uses new index to join 2 dataframe

# join

```python
df1.join(df2,index="Baseball")
```

```python
df1.set_index("Baseball")
df2.set_index("Baseball")
```

```python
df1.join(df2)
```

```python
df1.join(df2, lsuffix='_left', rsuffix='_right')
```

## Handling Missing values

# 1. dropna

```python
dict = {
    "Cricket" :[1,6,np.nan,4,3],
    "Football":[4,2,np.nan,5,np.nan],
    "Baseball":[2,4,5,7,1]
}
df= pd.DataFrame(dict)
df
```

```python
# Drop entire row that contain missing values
df.dropna()
```

```python
# drop entire column that contain missing values
df.dropna(axis=1)
```

```python
# drop entire row if it contain  than 1 value
#thresh takes integer value which tells minimum amount of na values to drop. 
df.dropna(thresh=2)
```

```python
df.dropna(thresh=1)
```

```python
df.dropna(thresh=3)
```

# 2. fillna

```python
# fill missing values with 0
df.fillna(value=0)
```

```python
# fill missing value with it's total mean value
df["Football"].fillna(df["Football"].mean())
```
