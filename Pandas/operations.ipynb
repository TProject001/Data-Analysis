# Indexing and Slicing

```python
from numpy.random import randn
np.random.seed(50)
df = pd.DataFrame(randn(10,5),index='A B C D E F G H I J'.split(),columns="Score1 Score2 Score3 Score4 Score5".split())
df
```

# Accessing Columns

```python
#way 1
df['Score1']
```

```python
df.Score3
```

# Accessing rows

```python
# way 1
df.loc["A"]
```

```python
df.loc['A':'D']
```

```python
df.loc["A","Score4"]
```

```python
df.loc[["D","I"],["Score3","Score5"]]
```

```python
# way 2
df.iloc[0]
```

```python
df.iloc[-1]
```

```python
df.iloc[:3] # last index is exclusive
```

```python
type(df['Score1'])
```

```python
df[["Score2",'Score1',"Score4"]]
```

# Creating new column

```python
df["Score6"] = df["Score5"] + df["Score3"]
df
```

# drop()

```python
# drop row
df.drop("Score6",axis=1)
df  # It does not show any changes because it create new data frame for changes
```

```python
x = df.drop("Score6",axis=1)
x
```

```python
x = df.drop("Score6",axis=0)
x # KeyError: "['Score6'] not found in axis"
```

```python
# To show the change in original dataframe 
df.drop("Score6",axis=1,inplace=True)
df
```

```python
# drop column
df.drop("A",axis=0,inplace=True)
df
```

# Conditional Selection

```python
# Similar to numpy we can use conditional selecting using bracket
# all values in dataframe grater than particular value (suppose >1.5)
df > 1.5 # It return boolean values
```

```python
df[df>1.5] # It return Numerical values
```

```python
df[df["Score3"] > 1.5] # It return rows (Apply condition to only Score3 column)
```

```python
df[df["Score3"] > 1.5]["Score3"]
```

```python
df[df["Score3"] > 1.5][["Score1","Score3","Score4"]]
# as Score4 does not have any value greater than 1.5 it return by default value present next to Score3
```
